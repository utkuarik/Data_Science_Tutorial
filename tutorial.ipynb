### 1- Import required libraries

```python
import pandas as pd
```

### 2-Import the datasets

```python
data = pd.read_csv('data.csv')
data[['Name','Value']].head()
```

### 3-Check the column names

```python
data.columns
```

### 4-Get some descriptive statistics

```python
data.describe()
```

#### Count the number of players by each country:

```python
data[['ID', 'Nationality']].groupby(['Nationality']).count().sort_values(by='ID', ascending=False).head(10)
```

#### Average overall score for each nationality:

```python
data['count'] = data.groupby('Nationality')['ID'].transform('count')
data.loc[data['count'] > 200].groupby(['Nationality'])[['Potential']].mean().sort_values(by='Potential', ascending= False)
```

#### Turkish players with highest potentials:

```python
data.loc[data['Nationality'] == 'Turkey', ['Name', 'Potential','Value']].sort_values(by='Potential', ascending= False).head(15)
```

#### English players with highest potentials:

```python
data.loc[data['Nationality'] == 'England', ['Name', 'Potential','Value']].sort_values(by='Potential', ascending= False).head(15)
data['Value'].head()
```

#### Convert 'Value' column to double data type:

```python
data['Value'] = data['Value'].apply(lambda x: x.replace("€", ""))  
data['Value'] = data['Value'].apply(lambda x: x.replace("M", "00000") if "." in x else x.replace("M", "000000"))
data['Value'] = data['Value'].apply(lambda x: x.replace("K", "000"))  
data['Value'] = data['Value'].apply(lambda x: x.replace(".", ""))  
data['Value'] = data['Value'].astype("double")
data['Value'].head()
```

```python
data[['Name','Value']].sort_values(by = 'Value', ascending= False).head(10).reset_index(drop=True)
```

```python
data[['Club','Value']].groupby('Club').agg({'Value':'sum'}).sort_values(by='Value', ascending=False).head(10)
```
